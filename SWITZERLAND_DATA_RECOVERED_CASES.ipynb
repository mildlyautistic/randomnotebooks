```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
from datetime import timedelta
```

```python
df = pd.read_csv('covid19_released_switzerland_openzh.csv')
```

```python
df.head()
```

```python
df.fillna( method ='ffill', inplace = True)
```

```python
df.fillna(0 , inplace = True)
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df["AG"],label="AG")
plt.plot(df["AI"],label="AI")
plt.plot(df["AR"],label="AR")
plt.plot(df["VD"],label="VD")
plt.plot(df["VS"],label="VS")
plt.plot(df["ZG"],label="ZG")

plt.ylabel("Number of Patients RELEASED IN KANTONS")
plt.xlabel("Date")
plt.legend()
plt.title("GROWTH RATE OF RECOVERED CASES IN KANTONS")
plt.xticks(rotation=90)
plt.show() 
```

```python
plt.figure(figsize=(20,7))
plt.plot(df["CH"],label="CH")

plt.ylabel("Number of Patients released")
plt.xlabel("Date")
plt.legend()
plt.title("GROWTH RATE OF RELEASED CASES IN SWITZERLAND DAY WISE")
plt.xticks(rotation=90)
plt.show() 
```
